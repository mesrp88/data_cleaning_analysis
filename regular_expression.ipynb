```python
import re
python_mentions=0
pattern="[Pp]ython"
titles = hacker_news["title"].tolist()
for title in titles:
    if re.search(pattern, title):
        python_mentions+=1
python_mentions
```

```python
pattern = '[Pp]ython'
titles= hacker_news["title"]
python_mentions=titles.str.contains(pattern).sum()
python_mentions
```

```python
titles = hacker_news['title']
ruby_titles=titles[titles.str.contains('[Rr]uby')]
ruby_titles
```

```python
email_bool= titles.str.contains('e[-]?mail')

email_count= email_bool.sum()
print(email_count)
email_titles= titles[email_bool]
email_titles.head(20)
```

```python
pattern= "\[\w+]"   #w represent any one character  + represent more and 
tag_titles= titles.str.contains(pattern)
tag_count = tag_titles.sum()
tag_count
```

```python
extract_value= titles.str.extract(r"(\[\w+\])", expand=False).value_counts()
extract_value
```

```python
titles.head()
```

```python
pattern = r"\[(\w+)\]"
tag_freq=titles.str.extract(pattern, expand=False).value_counts()
tag_freq
```

```python
def first_10_matches(pattern):
    """
    Return the first 10 story titles that match
    the provided regular expression
    """
    all_matches = titles[titles.str.contains(pattern)]
    first_10 = all_matches.head(10)
    return first_10
```

```python
first_10_matches(r"[Jj]ava[^Ss]")
```

```python
beginning_count= titles.str.contains(r"^\[\w+\]").sum()
print(beginning_count)
ending_count= titles.str.contains(r"\[\w+\]$").sum()
print(ending_count)
```
